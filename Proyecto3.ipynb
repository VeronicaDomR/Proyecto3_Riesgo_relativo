```python
# Autenticación
from google.colab import auth
auth.authenticate_user()

# Conectar a BigQuery e importar librerías
from google.cloud import bigquery
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np  # Usaremos numpy para el cálculo de z-scores

# Definir tu proyecto, dataset y tabla
project_id = "proyecto3-428922"
dataset = "dataset"
table = "default_user_info_union"

# Crear un cliente de BigQuery
client = bigquery.Client(project=project_id)

# Crear una consulta SQL para extraer los datos de la tabla
query = f"""
SELECT * FROM `{project_id}.{dataset}.{table}`
"""

# Ejecutar la consulta y cargar los datos en un DataFrame de pandas
df = client.query(query).to_dataframe()

# Calcular z-scores para las variables de interés
cols_of_interest = ['age']

# Calcular z-scores utilizando numpy
z_scores = (df[cols_of_interest] - df[cols_of_interest].mean()) / df[cols_of_interest].std()

# Definir umbral para outliers (por ejemplo, z-score > 3 o < -3)
outliers = df[(np.abs(z_scores) > 3).any(axis=1)]

# Visualizar outliers con un boxplot con colores personalizados
plt.figure(figsize=(10, 6))
box = df[cols_of_interest].boxplot(patch_artist=True, return_type='dict')

# Personalizar colores
box_color = '#6c14a5'
median_color = '#fcaf40'

# Personalizar las cajas
for box_element in box['boxes']:
    box_element.set_facecolor(box_color)

# Personalizar los bigotes
for whisker in box['whiskers']:
    whisker.set_color(box_color)

# Personalizar las tapas
for cap in box['caps']:
    cap.set_color(box_color)

# Personalizar las medianas
for median in box['medians']:
    median.set_color(median_color)

# Personalizar los puntos de los outliers
for flier in box['fliers']:
    flier.set(marker='o', color=box_color, alpha=0.5)

plt.title('Boxplot de Variables con Outliers (Z-Score)')
plt.ylabel('Valores')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Mostrar los datos de los outliers
print("Datos de Outliers:")
print(outliers)
```

```python
import pandas as pd
from tabulate import tabulate

# Cargar los datos en DataFrames desde los archivos subidos
user_info = pd.read_csv('/content/user_info.csv')
loans_detail = pd.read_csv('/content/loans_detail.csv')
default = pd.read_csv('/content/default.csv')
datset1 = pd.read_csv('/content/dataset1.csv')

# Combinar los DataFrames en uno solo para facilitar el análisis
combined_df = pd.merge(user_info, loans_detail, on='user_id', how='left')
combined_df = pd.merge(combined_df, default, on='user_id', how='left')

# Lista de las columnas de interés
columns_of_interest = [
    'age',
    'last_month_salary',
    'number_dependents',
    'more_90_days_overdue',
    'using_lines_not_secured_personal_assets',
    'number_times_delayed_payment_loan_30_59_days',
    'debt_ratio',
    'number_times_delayed_payment_loan_60_89_days',
    'default_flag'
]

# Filtrar solo las columnas de interés que están presentes en el DataFrame combinado
filtered_columns_of_interest = [col for col in columns_of_interest if col in combined_df.columns]
combined_df = combined_df[filtered_columns_of_interest]

# Calcular las estadísticas descriptivas
stats = combined_df.describe().transpose()

# Calcular la moda
mode_values = combined_df.mode().iloc[0]
stats['mode'] = mode_values

# Calcular la media y mediana
stats['mean'] = combined_df.mean()
stats['median'] = combined_df.median()

# Seleccionar columnas específicas para mostrar
result = stats[['min', 'max', 'mode', 'mean', 'median']]

# Imprimir el resultado como una tabla
print(tabulate(result, headers='keys', tablefmt='grid'))
```

```python
import pandas as pd

# Crear un DataFrame de ejemplo (asegúrate de usar tus datos reales)
data = {
    'more_90_days_overdue_status': ['Yes', 'No', 'No', 'Yes'],
    'Riesgo_Incumplimiento': ['High', 'Low', 'Medium', 'High'],
    'Capacidad_Endeudamiento': ['Bajo', 'Medio', 'Alto', 'Bajo'],
    'age': [22, 45, 37, 29]
}

df = pd.DataFrame(data)

# Crear variables dummies para 'more_90_days_overdue_status'
df = pd.get_dummies(df, columns=['more_90_days_overdue_status'], prefix='overdue', dummy_na=False)

# Crear variables dummies para 'Riesgo_Incumplimiento'
df = pd.get_dummies(df, columns=['Riesgo_Incumplimiento'], prefix='risk', dummy_na=False)

# Crear variables dummies para 'Capacidad_Endeudamiento'
df = pd.get_dummies(df, columns=['Capacidad_Endeudamiento'], prefix='debt_capacity', dummy_na=False)

# Crear la columna 'age_group' solo si 'age' es numérica
if pd.api.types.is_numeric_dtype(df['age']):
    # Categorizar 'age' en grupos
    df['age_group'] = pd.cut(df['age'], bins=[0, 20, 40, 60, 80, 100], labels=['0-20', '21-40', '41-60', '61-80', '81-100'])
    # Crear variables dummies para 'age_group'
    df = pd.get_dummies(df, columns=['age_group'], prefix='age', dummy_na=False)
else:
    # Crear variables dummies directamente en 'age' si ya es categórica
    df = pd.get_dummies(df, columns=['age'], prefix='age', dummy_na=False)

# Mostrar las primeras filas del DataFrame resultante
print(df.head())
```

```python
# Crear variables dummy para las columnas 'more_90_days_overdue_status' y 'Riesgo_Deuda'
# Asegúrate de que los nombres de las columnas coincidan con los nombres en el DataFrame
dummies = pd.get_dummies(score_risk[['generational_group', 'more_90_days_overdue_status', 'Riesgo_Deuda', 'payment_compliance_status']],
                         prefix=['Generational_group','More_90_days_overdue_status', 'Riesgo_deuda', 'Payment_compliance_status'])
# Reemplazar valores True por 1 y False por 0 (esto depende de cómo están representados los valores en tus datos)
dummies = dummies.replace({True: 1, False: 0})
# Concatenar las columnas dummy con el DataFrame original
score_risk_expanded = pd.concat([score_risk, dummies], axis=1)
# Mostrar el DataFrame resultante
display(score_risk_expanded)
# Variables predictoras
X = score_risk_expanded[['Generational_group_Millenials', 'Riesgo_deuda_Riesgo Medio', 'Riesgo_deuda_Bajo Riesgo', 'Riesgo_deuda_Alto Riesgo', 'More_90_days_overdue_status_Poco Retraso',
                         'More_90_days_overdue_status_Mucho Retraso', 'More_90_days_overdue_status_Medio Retraso', 'Payment_compliance_status_Cumplido',
                         'Payment_compliance_status_Incumplido', 'Payment_compliance_status_Intermedio']]
#

# Variable objetivo
y = score_risk_expanded['default_flag']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear y ajustar el modelo
model = LogisticRegression()
model.fit(X_train, y_train)

# Predecir en el conjunto de prueba
y_pred = model.predict(X_test)

# Evaluar el modelo
print(classification_report(y_test, y_pred))
print("ROC AUC Score:", roc_auc_score(y_test, model.predict_proba(X_test)[:,1]))

# Obtener las probabilidades de ser un cliente con riesgo crediticio
score_risk_expanded['predicted_prob'] = model.predict_proba(X)[:, 1]

# Mostrar las primeras filas del DataFrame con las probabilidades predichas
display(score_risk_expanded[['default_flag', 'predicted_prob']].head())

```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, roc_curve, RocCurveDisplay, roc_auc_score
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Definir las características y la variable objetivo
X = score_risk[['generational_group', 'more_90_days_overdue_status', 'Riesgo_Deuda', 'payment_compliance_status']]
y = score_risk['default_flag']
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Definir el preprocesador
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), ['generational_group', 'more_90_days_overdue_status', 'Riesgo_Deuda', 'payment_compliance_status'])
    ])
# Definir el modelo de regresión logística con balanceo de clases
logistic_model = LogisticRegression(class_weight='balanced')
# Definir el sobremuestreo SMOTE
smote = SMOTE(random_state=42)
# Crear el pipeline que incluye SMOTE, el preprocesador y el modelo
pipeline = ImbPipeline(steps=[
    ('preprocessor', preprocessor),
    ('smote', smote),
    ('model', logistic_model)
])
# Entrenar el modelo
pipeline.fit(X_train, y_train)
# Obtener los nombres de las características del preprocesador
feature_names = pipeline.named_steps['preprocessor'].get_feature_names_out()
# Obtener los coeficientes (w) y el intercepto (b) del modelo ajustado
w = pipeline.named_steps['model'].coef_[0]
b = pipeline.named_steps['model'].intercept_[0]
print(f'Coeficientes w: {w}')
print(f'Intercepto b: {b}')
# Hacer predicciones usando el modelo ajustado
y_pred = pipeline.predict(X_test)
# Calcular la precisión
accuracy = accuracy_score(y_test, y_pred)
print(f'Precisión: {accuracy:.2f}')
# Matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print(conf_matrix)
# Graficar la matriz de confusión
plt.figure(figsize=(8, 6))
ConfusionMatrixDisplay(conf_matrix, display_labels=['No Default', 'Default']).plot(cmap='Blues')
plt.title('Matriz de Confusión')
plt.show()
# Reporte de clasificación
class_report = classification_report(y_test, y_pred, zero_division=0)
print('Classification Report:')
print(class_report)
# Calcular y mostrar la curva ROC
fpr, tpr, _ = roc_curve(y_test, pipeline.predict_proba(X_test)[:, 1])
roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc_score(y_test, pipeline.predict_proba(X_test)[:, 1]))
roc_display.plot()
plt.title('Curva ROC')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
import pandas as pd

# Definir las características y la variable objetivo
X = score_risk[['payment_compliance_status']]
y = score_risk['default_flag']

# Convertir 'payment_compliance_status' a una variable numérica
le = LabelEncoder()
X['payment_compliance_status_encoded'] = le.fit_transform(X['payment_compliance_status'])

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X[['payment_compliance_status_encoded']], y, test_size=0.3, random_state=42)

# Definir el modelo de regresión logística
logistic_regression = LogisticRegression(max_iter=1000, random_state=42)

# Entrenar el modelo
logistic_regression.fit(X_train, y_train)

# Predecir probabilidades
y_pred_prob = logistic_regression.predict_proba(X_test)[:, 1]

# Crear un DataFrame con las características originales y las probabilidades predichas
X_test_with_prob = X_test.copy()
X_test_with_prob['pred_prob'] = y_pred_prob

# Crear los puntos para la curva logística
x_values = np.linspace(X['payment_compliance_status_encoded'].min(), X['payment_compliance_status_encoded'].max(), 100)
y_values = logistic_regression.predict_proba(x_values.reshape(-1, 1))[:, 1]

# Graficar la probabilidad predicha
plt.figure(figsize=(10, 6))
plt.scatter(X_test_with_prob['payment_compliance_status_encoded'], y_test, label='Datos Reales', alpha=0.5)
plt.plot(x_values, y_values, color='red', label='Curva de Regresión Logística')
plt.xlabel('Payment Compliance Status (Encoded)')
plt.ylabel('Probability of Default')
plt.title('Logistic Regression Curve')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

train = score_risk
train['var_temp'] = train['generational_group'].map({'Baby Boomers': 1, 'Generacion X': 2,  'Millenials': 3,  'Generacion Z': 4})

X = train[['var_temp']].values  # Variable predictora
y = train['default_flag'].values  # Variable objetivo
# Crear y ajustar el modelo de regresión logística con class_weight='balanced'
model = LogisticRegression(class_weight='balanced')
model.fit(X, y)
# Obtener los coeficientes (w) y el intercepto (b) del modelo ajustado
w = model.coef_[0][0]
b = model.intercept_[0]
print(f'Coeficiente w: {w}')
print(f'Intercepto b: {b}')
# Hacer predicciones usando el modelo ajustado
y_pred = model.predict(X)
# Calcular la precisión
accuracy = accuracy_score(y, y_pred)
print(f'Precisión: {accuracy:.2f}')
# Matriz de confusión
conf_matrix = confusion_matrix(y, y_pred)
print('Confusion Matrix:')
print(conf_matrix)
# Graficar la matriz de confusión
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicción 0', 'Predicción 1'],
            yticklabels=['Real 0', 'Real 1'])
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.title('Matriz de Confusión')
plt.show()
# Reporte de clasificación
class_report = classification_report(y, y_pred, zero_division=0)
print('Classification Report:')
print(class_report)
# Puntos de la recta
x = np.linspace(0, train['var_temp'].max(), 100)
y_logistic = 1 / (1 + np.exp(-(w * x + b)))
# Graficar la función logística
plt.figure(figsize=(8, 6))
plt.scatter(train['var_temp'], y, color='blue', label='Datos Reales')
# Graficar la función logística
plt.plot(x, y_logistic, '-r', label='Función Logística')
plt.xlabel('var_tempa')
plt.ylabel('Probabilidad')
plt.title('Regresión Logística')
plt.ylim(0, 1.1)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
import pandas as pd

# Crear un DataFrame de ejemplo (puedes omitir esta parte si ya tienes tus datos)
data = {
    'generational_group': ['Millennials', 'Gen X', 'Boomers', 'Gen Z', 'Gen X', 'Boomers'],
    'more_90_days_overdue_status': ['Poco Retraso', 'Mucho Retraso', 'Medio Retraso', 'Poco Retraso', 'Mucho Retraso', 'Medio Retraso'],
    'Riesgo_Deuda': ['Alto Riesgo', 'Bajo Riesgo', 'Riesgo Medio', 'Alto Riesgo', 'Bajo Riesgo', 'Riesgo Medio'],
    'payment_compliance_status': ['Cumplido', 'Incumplido', 'Intermedio', 'Cumplido', 'Incumplido', 'Intermedio'],
    'default_flag': [1, 0, 1, 0, 1, 0]
}

score_risk = pd.DataFrame(data)

# Definir las características y la variable objetivo
X = score_risk[['generational_group', 'more_90_days_overdue_status', 'Riesgo_Deuda', 'payment_compliance_status']]
y = score_risk['default_flag']

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Definir el preprocesamiento para variables categóricas con manejo de categorías desconocidas
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), ['generational_group', 'more_90_days_overdue_status', 'Riesgo_Deuda', 'payment_compliance_status'])
    ])

# Crear el pipeline con el preprocesador y el modelo
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression())
])

# Entrenar el modelo
model.fit(X_train, y_train)

# Predecir las probabilidades de incumplimiento en el conjunto de prueba
score_crediticio = model.predict_proba(X_test)[:, 1]  # Probabilidad de default
score_risk_test = X_test.copy()
score_risk_test['default_flag'] = y_test
score_risk_test['score_crediticio'] = 1 - score_crediticio  # Invertir la probabilidad para que un score más alto signifique menor riesgo

# Normalizar el score a una escala de 0 a 100
score_risk_test['score_crediticio'] = score_risk_test['score_crediticio'] * 100

# Calcular el promedio del score crediticio
promedio_score_crediticio = score_risk_test['score_crediticio'].mean()
print("Promedio del score crediticio:", promedio_score_crediticio)

# Visualización del score crediticio
plt.figure(figsize=(10, 6))
sns.histplot(score_risk_test['score_crediticio'], bins=10, kde=True)
plt.title('Distribución del Score Crediticio')
plt.xlabel('Score Crediticio')
plt.ylabel('Frecuencia')
plt.show()

```
